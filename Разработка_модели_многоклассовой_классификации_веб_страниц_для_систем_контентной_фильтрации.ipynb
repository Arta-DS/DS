# Тема проекта: "Разработка модели многоклассовой классификации веб-страниц для систем контентной фильтрации".

**Актуальность и постановка задачи:**

**Проблема**:

В современных корпоративных сетях и для родительского контроля остро стоит задача ограничения доступа к определенным категориям сайтов (соцсети, азартные игры и т.д.). Ручная категоризация миллионов новых доменов в день невозможна. Существующие решения требуют постоянного обновления и могут быть неточными.

**Место в предметной области:**

Проект будет упрощенной версией промышленных систем, таких как WebCat от Solar Security. На основе текстового содержимого страницы будет автоматически присваиваться ей одну из заданных категорий.

**Цель:**

Создать модель машинного обучения, которая по текстовому содержимому веб-страницы (очищенному HTML-коду) с высокой точностью определяет ее категорию.

**Бизнес-метрики:**

В контентной фильтрации цена ошибки очень высока, и разная для разных типов ошибок.

False Positive (FP): Сайт "Новости спорта" ошибочно заблокирован как "Соцсети". Это приводит к недовольству легитимных пользователей и обращениям в поддержку.
False Negative (FN): Сайт "Онлайн-казино" пропущен фильтром как "Новости". Это прямая угроза политике безопасности.
Поэтому простая Accuracy нам не подходит. Нам нужен баланс.

**Основная техническая метрика:**

Macro F1-score > 0.90.
Почему Macro F1? Эта метрика вычисляет F1-меру для каждого класса, а затем усредняет их. Это заставляет модель хорошо работать на всех категориях, включая редкие, но критичные (например, "Взрослый контент"), а не только на самых частых.

**Итоговая бизнес-метрика:**

Снижение количества ложных срабатываний (FP + FN) на 20% по сравнению с базовой эвристической системой.
Как это измерить в проекте: В реальном мире это измеряется по обращениям в поддержку. В рамках проекта мы сформулируем это так: "Достижение высокого Macro F1-score напрямую ведет к минимизации как FP, так и FN, что является ключевой целью для любой системы контентной фильтрации".
